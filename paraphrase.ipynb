```python
from sentence_transformers.util import paraphrase_mining
paraphrases = paraphrase_mining(model,sentences)
```

```python
for paraphrase in paraphrases:
    score,i,j = paraphrase
    print("{}\t\t\t\t\t {} \t\t score : {:.2f}".format(sentences[i],sentences[j],score))
```

```python
import pandas as pd

data = []

for paraphrase in paraphrases:
    score, i, j = paraphrase
    data.append({
        "Sentence 1": sentences[i],
        "Sentence 2": sentences[j],
        "Score": round(score, 2)
    })

df = pd.DataFrame(data)
df
```
